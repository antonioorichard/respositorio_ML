# Os par√¢metros do Affinity Propagation

1. Par√¢metros Affinity Propagation
* max_iter: N√∫mero m√°ximo de itera√ß√µes. O valor padr√£o de f√°brica √© 200 convergence_iter:
N√∫mero de itera√ß√µes que interrompe o treinamento, quando n√£o houver mais mudan√ßas no n√∫mero estimado de cluster
* copy: Faz uma c√≥pia dos dados de entrada
* preference: O par√¢metro "preference" controla a prioridade ou prefer√™ncia que os pontos t√™m em se tornar centros de cluster. Valore
* damping: O fator de amortecimento no intervalo [0,5, 1,0) controla a taxa de atualiza√ß√£o das responsabilidades e disponibilidade
* affinity: Afinidade √© a m√©trica de similaridade entre os pontos do conjunto de dados. No momento, s√£o suportados 'precomputed
* random_state: Gerador de n√∫meros pseudo-aleat√≥rios para controlar o estado inicial. Use um int para resultados reproduz√≠veis em cha

## Convergencia do Affinity Propagation

# 1. Converg√™ncia do algoritmo

```python
from matplotlib import pyplot as plt
from sklearn import datasets as dt
from sklearn import cluster as c
import numpy as np
from sklearn import metrics as m
import time
from IPython import display
# dataset
X, clusters = dt.make_blobs( n_samples=300, centers=4, cluster_std=0.60,
random_state=0)
# show dataset
plt.scatter(X[:,0], X[:,1], alpha=0.7, edgecolors='b')
```

```python
# Convergency testing
p = np.arange( -1, -61, -1 )
ss_list = []
fig = plt.figure()
for i in range( len( p ) ):
    model = c.AffinityPropagation( preference=p[i] )
    model.fit( X )
    labels = model.predict( X )
    # Metric
    ss = m.silhouette_score( X, labels )
    ss_list.append( ss )
    # Plot
    plt.plot( p[:i], ss_list[:i] )
    # Plot
    plt.title( 'Affinity Propagation - Convergence' );
    plt.xlabel( 'Preference' );
    plt.ylabel( 'Silhouette Score' )
    display.clear_output(wait=True)
    plt.show()
    plt.gcf()
    time.sleep(0.05)
```

```python
# Showing final Clustering
max_index = ss_list.index( max( ss_list ) )
af = c.AffinityPropagation( preference=p[max_index] )
clustering = af.fit(X)
fig = plt.figure()
plt.title( "Separa√ß√£o dos Clients VIP")
plt.scatter(X[:,0], X[:,1], c=clustering.labels_, cmap='rainbow', alpha=0.7, edgecolors='b');
```
